```python
df = pd.read_csv('data/stem_female.csv')
```

```python
df = df.drop(['Unnamed: 0', 'Country Name', 'Country Code', 'Indicator Name',
       'Indicator Code'], axis = 1)
```

```python
df[['2015', '2016', '2017']]
```

```python
pd.DataFrame(df.median(axis = 1).sort_values(ascending = False))
```

```python
top_20 = pd.DataFrame(df[['2015','2016','2017']].mean(axis = 1).sort_values(ascending = False)[:20], columns = ['median'])
all_ = pd.DataFrame(df[['2015','2016','2017']].mean(axis = 1).sort_values(ascending = False), columns = ['median'])
```

```python
top_20
```

```python
plt.figure(figsize=(12, 8))
splot=sns.barplot(y = top_20.index, x = 'median', data=top_20, color="skyblue")
# plt.barh(y = top_20.index, x = 'median', data = top_20, width  )
plt.xlabel("", size=14)
plt.xticks([])
plt.ylabel("Country", size=14)
plt.title('Share of female STEM graduates by Country')
# for p in splot.patches:
#     splot.annotate(format(p.get_width(), '.1f'), 
#                    (p.get_x() + p.get_height() / 2., p.get_width()), 
#                    ha = 'center', va = 'center', 
#                    size=9,
#                    xytext = (0, -12), 
#                    textcoords = 'offset points')
for rect in splot.patches:
    width = rect.get_width()
    plt.text(0.975*rect.get_width(), rect.get_y()+0.5*rect.get_height(),
             '%d' % int(width),
             ha='center', va='center')
# plt.bar_label(splot.containers[0],size=16,label_type='center')
# plt.savefig("add_annotation_to_bars_in_barplot_Seaborn_Python.png")
```
