# Exploratory Data Analysis

```python
# import relevant libaries 
import numpy as np 
import pandas as pd 

# Data Visualization
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
stem = pd.read_csv('data/stem_female.csv')
```

```python
stem.head()
```

```python
stem.shape
```

```python
stem.isna().sum()
```

```python
stem.index = stem['Country Name']
```

```python
stem.head()
```

```python
stem = stem.drop(['Country Name', 'Country Code', 'Indicator Name',
       'Indicator Code'], axis = 1)
```

```python
stem[['2015', '2016', '2017']].head()
```

```python
pd.DataFrame(stem.median(axis = 1).sort_values(ascending = False))
```

```python
top_20 = pd.DataFrame(stem[['2015','2016','2017']].mean(axis = 1).sort_values(ascending = False)[:20], columns = ['central'])
bottom_20 = pd.DataFrame(stem[['2015','2016','2017']].mean(axis = 1).sort_values(ascending = True)[:20], columns = ['central'])
all_ = pd.DataFrame(stem[['2015','2016','2017']].mean(axis = 1).sort_values(ascending = False), columns = ['central'])
```

```python
bottom_20
```

```python
plt.figure(figsize=(12, 8))
splot=sns.barplot(y = top_20.index, x = 'central', data=top_20, color="skyblue")
plt.xlabel("", size=14)
plt.xticks([])
plt.ylabel("Country", size=14)
plt.title('Share of female STEM graduates by Country (Top 20)')
for rect in splot.patches:
    width = rect.get_width()
    plt.text(0.975*rect.get_width(), rect.get_y()+0.5*rect.get_height(),
             '%d' % int(width),
             ha='center', va='center')
# plt.bar_label(splot.containers[0],size=16,label_type='center')
# plt.savefig("add_annotation_to_bars_in_barplot_Seaborn_Python.png")
```

```python
plt.figure(figsize=(12, 8))
splot=sns.barplot(y = bottom_20.index, x = 'central', data=bottom_20, color="skyblue")
plt.xlabel("", size=14)
plt.xticks([])
plt.ylabel("Country", size=14)
plt.title('Share of female STEM graduates by Country (Bottom 20)')
for rect in splot.patches:
    width = rect.get_width()
    plt.text(0.975*rect.get_width(), rect.get_y()+0.5*rect.get_height(),
             '%d' % int(width),
             ha='center', va='center')
# plt.bar_label(splot.containers[0],size=16,label_type='center')
# plt.savefig("add_annotation_to_bars_in_barplot_Seaborn_Python.png")
```

```python
# boxplot
plt.figure(figsize=(12, 8))
splot = sns.barplot(x = migrant_mean.index, y = 'Percentage', data=migrant_mean, color = 'skyblue')
plt.xlabel("Year", size=14)
# plt.xticks([])
plt.ylabel("Mean Share", size=14)
plt.title('Mean Share of Female Migrants', size = 16)
for rect in splot.patches:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2., height,
            '%.1f' % float(height),
            ha='center', va='bottom')
# for p in splot.patches:
#     splot.annotate(format(p.get_height(), '.1f'), 
#                    (p.get_x() + p.get_width() / 2., p.get_height()), 
#                    ha = 'center', va = 'center', 
#                    size=15,
#                    xytext = (0, -12), 
#                    textcoords = 'offset points')
# plt.savefig("share_of_female_migrants.png")
```

```python
regions = ['Latin America & Caribbean', 'South Asia', 'Sub-Saharan Africa',
           'Europe & Central Asia', 'Middle East & North Africa',
           'East Asia & Pacific', 'North America']
region_df = migrant.copy()
# region_df.reset_index(inplace = True, drop = True),
region_df
```

```python
region_df
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data = region_df, x = 'Year', y = 'Percentage', hue = 'Country Name')
plt.xlabel("Year", size=14)
# plt.xticks([])
plt.ylabel("Share (%)", size=14)
plt.title('Percentage of Female Migrants by region', size = 16)
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
# plt.savefig("female_migrants_by_region.png")
```

```python
income_df = pd.read_csv('data/female_migrant_income.csv')
income_df.head()
```

```python
income_df = income_df.groupby('IncomeGroup').mean()
# income_df
income_df.reset_index(inplace = True)
```

```python
income_df = income_df.melt(id_vars='IncomeGroup',
                   var_name='Year',
                   value_name='Percentage')
```

```python
income_df
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data = income_df, x = 'Year', y = 'Percentage', hue = 'IncomeGroup', color = 'skyblue')
plt.xlabel("Year", size=14)
# plt.xticks([])
plt.ylabel("Share (%)", size=14)
plt.title('Percentage of Female Migrants by region', size = 16)
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
# plt.savefig("female_migrants_by_region.png")
```
